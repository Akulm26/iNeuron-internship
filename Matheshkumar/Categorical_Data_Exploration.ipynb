```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

pd.set_option("display.max_rows", None, "display.max_columns", None)
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv("Train.csv")
df.head(2)
```

```python
[i for i in df.columns if df.dtypes[i]=='object']
```

```python
# removing spelling error in variables
itemFatContent = {'low fat': "Low Fat", 'LF': "Low Fat", 'reg': "Regular"}
df.Item_Fat_Content.replace(itemFatContent, inplace=True)
```

```python
g1 = sns.countplot(x="Item_Fat_Content", data=df)

g1.bar_label(g1.containers[0])
```

```python
temp = pd.DataFrame()
temp['value_counts'] = df.Item_Fat_Content.value_counts()
temp['value_counts_percent'] = df.Item_Fat_Content.value_counts()/len(df)
temp
```
